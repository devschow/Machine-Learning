```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import sklearn 
%matplotlib inline
```

Load the dataset


```python
columns = ['Sepal length', 'Sepal Width', 'Petal lemgth', 'Petal width', 'Class labels']
df = pd.read_csv('iris.data', names = columns)
df.head() # Top 5 entries of the data
```

Data viz

```python
# describe gives stat analysis of your data
df.describe()
```

```python
# Viz the whole dataset
sns.pairplot(df,hue='Class labels') 
# Hue assigns differnt colors for distinct class
```

Initially the data is in a matrix form, We have to separate the data (input and output columns)
In this case separate features and target

```python
data = df.values

# Slicing operation
X = data[:,0:4] 
Y = data[:,4] 
print(Y)
```

Split Data for Training and Testing

```python
from sklearn.model_selection import train_test_split

# Test_size is given as 0.2 , it means 20% of our data goes into our test size
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.2) 

print(X_train)
```

**MODEL 1**: Support vector machine algorithm

```python
from sklearn.svm import SVC

#Model fitting = Training our system
model_svc = SVC()
model_svc.fit(X_train, y_train)
```

PREDICTION 

```python
prediction1 = model_svc.predict(X_test)
#Calculate accuracy

from sklearn.metrics import accuracy_score
print(accuracy_score(y_test, prediction1)* 100)
```

**MODEL 2**: Logistic Regression

```python
from sklearn.linear_model import LogisticRegression

#Model fitting = Training our system
modelLR = LogisticRegression(solver='lbfgs', max_iter=200)
modelLR.fit(X_train, y_train)
```

PREDICTION

```python
prediction2 = modelLR.predict(X_test)

#Calculate accuracy

from sklearn.metrics import accuracy_score
print(accuracy_score(y_test, prediction1)* 100)
```

```python
from sklearn.tree import DecisionTreeClassifier

model_DTC = DecisionTreeClassifier()
model_DTC.fit(X_train, y_train)
```
